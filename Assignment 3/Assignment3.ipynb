```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("cereal.csv")
```

```python
print(data.head())
```

```python
data.info()
```

```python
#summary stats
print(data.describe(include='all'))
```

```python
#bar chart to count by manufacturer, type, and shelf
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))
sns.countplot(data=data, x='mfr', palette="viridis")
plt.title("Count Plot for Manufacturer (mfr)")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(15, 5))
sns.countplot(data=data, x='type', palette="viridis")
plt.title("Count Plot for Type")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(15, 5))
sns.countplot(data=data, x='shelf', palette="viridis")
plt.title("Count Plot for Shelf")
plt.xticks(rotation=45)
plt.show()
```

```python
#box plot for calories by manufacturer (mfr), Type, and Shelf
plt.figure(figsize=(15, 5))
sns.boxplot(data=data, x='mfr', y='calories', palette="viridis")
plt.title("Box Plot of Calories by Manufacturer (mfr)")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(data=data, x='type', y='calories', palette="viridis")
plt.title("Box Plot of Calories by Type")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(15, 5))
sns.boxplot(data=data, x='shelf', y='calories', palette="viridis")
plt.title("Box Plot of Calories by Shelf")
plt.xticks(rotation=45)
plt.show()
```

```python
#violin for rating by manufacturer (mfr), type, and shelf
plt.figure(figsize=(15, 5))
sns.violinplot(data=data, x='mfr', y='rating', palette="viridis")
plt.title("Violin Plot of Rating by Manufacturer (mfr)")
plt.xticks(rotation=45)
plt.show()

data['type'] = data['type'].replace({'C': 'cold', 'H': 'hot'})

#color change
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='type', y='rating', palette={'cold': 'blue', 'hot': 'red'})
plt.title("Violin Plot of Rating by Type")
plt.xticks(rotation=45)
plt.ylabel("Rating")
plt.xlabel("Type")
plt.show()

plt.figure(figsize=(15, 5))
sns.violinplot(data=data, x='shelf', y='rating', palette="viridis")
plt.title("Violin Plot of Rating by Shelf")
plt.xticks(rotation=45)
plt.show()
```

```python
#stacked bar plot showing cereal by mfer and type (hot/cold) v. frequnecy 
cross_tab = pd.crosstab(data['mfr'], data['type'])

#designating hot and cold
cross_tab = cross_tab.rename(columns={'C': 'cold', 'H': 'hot'})

colors = ['blue', 'red']

cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors)
plt.title("Stacked Bar Plot of Manufacturer (mfr) vs Type")
plt.xlabel("Manufacturer (mfr)")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.legend(title="Type", loc='upper right')
plt.show()

```
