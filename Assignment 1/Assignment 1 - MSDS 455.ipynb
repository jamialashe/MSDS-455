## **Temporal Charts**

### **Line Graph**

```python
data = pd.read_csv('WalmartSales_cleandf.csv')

#remove whitespace
data.columns = data.columns.str.strip()

#convert date
data['Date'] = pd.to_datetime(data['Date'])

#date index
data.set_index('Date', inplace=True)

#average sales 
monthly_sales = data['Weekly_Sales'].resample('M').mean()  

#show monthly figures 
print(monthly_sales)

#plot
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales.index, monthly_sales.values, linewidth=2, marker = "o")
plt.title('Average Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Average Monthly Sales (Hundred Thousandths)')
plt.ylim(700000, 1500000)
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()


plt.show()
```

```python
#heatmap version, for stores 1-10
df = pd.read_csv('WalmartSales_cleandf.csv')
df['Date'] = pd.to_datetime(df['Date'])
filtered_df = df[(df['Store'].isin(range(1, 11))) & (df['Date'].dt.year == 2010)]
filtered_df['Month'] = filtered_df['Date'].dt.to_period('M')
monthly_sales = filtered_df.groupby(['Store', 'Month'])['Weekly_Sales'].mean().reset_index()
print(monthly_sales)
pivot_table = monthly_sales.pivot(index='Month', columns='Store', values='Weekly_Sales').fillna(0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='pink', annot=True, fmt=".1f", linewidths=0.5)

plt.title('Average Monthly Sales for Stores 1-10 (Year 2010)', fontsize=16)
plt.xlabel('Store', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#stacked area chart for storess 12-20
df = pd.read_csv('WalmartSales_cleandf.csv')

df['Date'] = pd.to_datetime(df['Date'])

filtered_df = df[(df['Store'].isin(range(12, 20))) & (df['Date'].dt.year == 2010)]

filtered_df['Month'] = filtered_df['Date'].dt.to_period('M')

#averages
monthly_sales = filtered_df.groupby(['Store', 'Month'])['Weekly_Sales'].mean().reset_index()

print(monthly_sales)

#pivot tables
pivot_table = monthly_sales.pivot(index='Month', columns='Store', values='Weekly_Sales').fillna(0)

#plot
plt.figure(figsize=(14, 7))
plt.stackplot(pivot_table.index.astype(str), pivot_table.T, labels=pivot_table.columns)
plt.title('Average Monthly Sales for Stores 1-10 (Year 2010)')
plt.xlabel('Month')
plt.ylabel('Average Monthly Sales (Hundred Thousdanths')
plt.legend(loc='upper left', title='Stores')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
